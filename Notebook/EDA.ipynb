```python
# Data Overview
import matplotlib.pyplot as plt
import numpy as np



class Data_overview:

    def __init__(self,data):
        self.data=data


    def clean_column_labels(self):
        """
        Remove spaces and hyphens from column labels and replace them with underscores in a Pandas DataFrame.

        Parameters:
        - dataframe: The input Pandas DataFrame.

        Returns:
        - A DataFrame with cleaned column labels.
        """
        dataframe=self.data

        # Use list comprehension to clean the column labels
        cleaned_columns = [col.replace(" ", "_").replace("-", "_") for col in dataframe.columns]
        
        # Rename the columns in the DataFrame
        dataframe.columns = cleaned_columns
        return dataframe

    def plot_unique_percentage(self):
        """
        Plot the percentage of unique values for each column in a DataFrame.

        Parameters:
        - dataframe: The input Pandas DataFrame.

        Returns:
        - None
        """
        # Accessing Dataframe
        dataframe=self.data
        


        unique_percentages = []
        column_names = []

        for column in dataframe.columns:
            unique_count = dataframe[column].nunique()
            total_count = len(dataframe)
            percentage = (unique_count / total_count) * 100

            unique_percentages.append(percentage)
            column_names.append(column)

        plt.figure(figsize=(10, 6))
        plt.bar(column_names, unique_percentages)
        plt.xlabel("Columns")
        plt.ylabel("Unique Value Percentage")
        plt.title("Unique Value Percentage for DataFrame Columns")
        plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
        plt.show()

        
    def plot_nan_percentage(self):
        """
        Plot the percentage of NaN values for each column in the Pandas DataFrame.

        Returns:
        - None
        """
        nan_percentages = (self.data.isna().mean() * 100)
        
        plt.figure(figsize=(10, 6))
        nan_percentages.plot(kind='bar')
        plt.xlabel("Columns")
        plt.ylabel("NaN Value Percentage")
        plt.title("NaN Value Percentage for DataFrame Columns")
        plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
        plt.show()


    def find_duplicate_rows(self):
        """
        Find and return duplicate rows in a Pandas DataFrame.

        Parameters:
        - dataframe: The input Pandas DataFrame.

        Returns:
        - A DataFrame containing the duplicate rows.
        """
        dataframe=self.data
        duplicate_rows = dataframe[dataframe.duplicated(keep='first')]
        return duplicate_rows
```

## Data Anlaysis 

```python
class data_analysis:
    def __init__(self,data):
        self.data=data
        pass
    def categorize_columns(self):
        """
        Identify and separate numerical and categorical columns in a Pandas DataFrame.

        Parameters:
        - dataframe: The input Pandas DataFrame.

        Returns:
        - Two lists: one for numerical columns and one for categorical columns.
        """
        dataframe=self.data

        numerical_columns = []
        categorical_columns = []

        for column in dataframe.columns:
            if np.issubdtype(dataframe[column].dtype, np.number):
                numerical_columns.append(column)
            else:
                categorical_columns.append(column)

        return numerical_columns, categorical_columns
    
    def plot_discrete_histogram_and_box(self, column_label, xlabel="Values", ylabel="Frequency", title="Histogram and Box Plot"):
        """
        Plot a histogram and a simple box plot for discrete data in a specific column of a Pandas DataFrame side by side.

        Parameters:
        - dataframe: The input Pandas DataFrame.
        - column_label: The label of the column to plot.
        - xlabel: Label for the x-axis (optional, default is "Values").
        - ylabel: Label for the y-axis (optional, default is "Frequency").
        - title: Title for the plot (optional, default is "Histogram and Box Plot").

        Returns:
        - None
        """

        dataframe=self.data


        if column_label not in dataframe.columns:
            print(f"Column '{column_label}' not found in the DataFrame.")
            return

        data = dataframe[column_label]

        fig, axes = plt.subplots(1, 2, figsize=(12, 6))

        # Plot histogram
        axes[0].hist(data, bins=len(set(data)), align='left', rwidth=0.8)
        axes[0].set_xlabel(xlabel)
        axes[0].set_ylabel(ylabel)
        axes[0].set_title("Histogram")

        # Plot simple box plot
        axes[1].boxplot(data)
        axes[1].set_xticklabels([column_label])
        axes[1].set_title("Box Plot")

        fig.suptitle(title, y=1.02)
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        plt.show()

    def plot_histogram_for_column(self, column_label, xlabel="Values", ylabel="Frequency", title="Histogram"):
        """
        Plot a histogram for a specific column in a Pandas DataFrame.

        Parameters:
        - dataframe: The input Pandas DataFrame.
        - column_label: The label of the column to plot.
        - xlabel: Label for the x-axis (optional, default is "Values").
        - ylabel: Label for the y-axis (optional, default is "Frequency").
        - title: Title for the plot (optional, default is "Histogram").

        Returns:
        - None
        """

        dataframe=self.data

        if column_label not in dataframe.columns:
            print(f"Column '{column_label}' not found in the DataFrame.")
            return

        data = dataframe[column_label]

        plt.figure(figsize=(8, 6))
        plt.hist(data, bins=len(set(data)), align='left', rwidth=0.8)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.title(title)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

```
